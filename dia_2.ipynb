```python
"""Challenge 🤺 - 20m

Escribe un programa que pregunte al usuario su edad para ingresar al festival Asunciónico y determine qué tipo de acceso puede tener.

Si el usuario tiene menos de 18 imprimir: "Acceso denegado — Solo zona Kids 🎈"

Si el usuario tiene mas de 18 pero menos de 21 imprimir: "Acceso permitido — No puede comprar bebidas alcoholicas 🎟️"

Si el usuario tiene mas de 21 imprimir: "Acceso total — Disfruta el festival 🎉"""


def Asuncionico():
    edadUser = int(input("Ingrese su edad"))
    if edadUser < 18:
        print("Acceso denegado — Solo zona Kids 🎈")
    elif edadUser < 21:
        print("Acceso permitido — No puede comprar bebidas alcoholicas 🎟️")
    elif edadUser > 21:
        print("Acceso total — Disfruta el festival 🎉")

Asuncionico()

```

```python
"""Challenge 🤺 - 30m
Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:

Entre -10 y 13 grados, que imprima "Ro'y koko".
Entre 14 y 23 grados, que imprima "Refresca'i".
Entre 24 y 28 grados, que imprima "Oikoite".
Entre 29 y 45 grados, que imprima "Hakurooooo".
Para todo lo demás, imprimir "Ñamano mba'e!"."""

def temperatura():
    temperaturaUser = int(input("Inserte la temperatura porfavor: "))
    if temperaturaUser >= -10 and temperaturaUser <= 13:
        print(f"Ro'y koko {temperaturaUser} grados")
    elif temperaturaUser >= 14 and temperaturaUser <= 23:
        print(f"Refresca'i {temperaturaUser} grados")
    elif temperaturaUser >= 24 and temperaturaUser <= 28:
        print(f"Oikoite {temperaturaUser} grados")
    elif temperaturaUser >= 29  and temperaturaUser <= 45:
        print(f"Hakurooooo {temperaturaUser} grados") 
    else:
        print(f"Ñamano mba'e! {temperaturaUser} grados")

temperatura()
```
